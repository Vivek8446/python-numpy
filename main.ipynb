### 2D array

```python
arr2 = np.array([[1,2],[3,4]])
print(arr2)
print(arr2.shape) #check the shape of the array
print(arr2[1,1])  #access the element at row 0 and column
```

## Zeros, Ones, Random

```python
print(np.zeros((2,3)))
print(np.ones((2,2)))
print(np.random.rand(2,2))
```

```python
a = np.array([1,2,3])
b = np.array([4,5,6])

print(a + b)   # [5 7 9]
print(a * 2)   # [2 4 6]
```

## maths functions

```python
arr = np.array([1,2,3,4])

print(np.mean(arr)) # mean of the array
print(np.sum(arr)) # sum of the array
print(np.max(arr)) # maximum value in the array
```

## matrix multiplication

```python
a = np.array([[1,2],[3,4]])
b = np.array([[5,6],[7,8]])

print(np.dot(a,b))
```

## reshape the array 

```python
arr = np.array([1,2,3,4,5,6])

new_arr = arr.reshape(2,3)
print(new_arr)
```
