## What is Pandas Actually Used For?

### Used for:

- Reading CSV / Excel

- Cleaning data

- Filtering

- Grouping

- Feature engineering

- Preparing data for ML

### If you build:

- Recommendation system

- Fraud detection

- Analytics dashboard

You WILL use Pandas.

```python
import pandas as pd

data = {
    "name": ["Vivek", "Rahul", "Aman"],
    "age": [22, 25, 21],
    "salary": [30000, 50000, 40000]
}

df = pd.DataFrame(data)
print(df)
```

```python
df.head()        # first 5 rows
df.tail()        # last 5 rows
df.info()        # structure
df.describe()    # statistics
df.shape         # rows, columns
df.columns       # column names
```
