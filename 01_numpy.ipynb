1ï¸âƒ£ Create 3x3 matrix with numbers 1â€“9
2ï¸âƒ£ Find row sum
3ï¸âƒ£ Find column sum
4ï¸âƒ£ Transpose matrix

```python
arr = np.arange(1, 10).reshape(3,3)

print("Matrix:\n", arr)

# Row sum
print("Row sum:", arr.sum(axis=1))

# Column sum
print("Column sum:", arr.sum(axis=0))

# Transpose
print("Transpose:\n", arr.T)
```

2ï¸âƒ£ Normalize marks (divide by 100)
3ï¸âƒ£ Replace marks below 35 with 0

```python
marks = np.array([40, 55, 78, 90, 32, 67, 30, 85])

# Above 60
print("Above 60:", marks[marks > 60])

# Normalize
print("Normalized:", marks / 100)

# Replace below 35
marks[marks < 35] = 0
print("Updated Marks:", marks)
```

ðŸ”¥ What is axis in NumPy?

Axis tells NumPy:

ðŸ‘‰ Which direction to perform operation.

ðŸ§  Example Matrix
import numpy as np

arr = np.array([[1,2,3],
                [4,5,6]])

Shape:

(2 rows, 3 columns)
ðŸ”¹ axis = 0 â†’ Down the rows (column-wise operation)
print(arr.sum(axis=0))

Output:

[5 7 9]

Because:

Column 1: 1 + 4 = 5
Column 2: 2 + 5 = 7
Column 3: 3 + 6 = 9
ðŸ”¹ axis = 1 â†’ Across columns (row-wise operation)
print(arr.sum(axis=1))

Output:

[ 6 15]

Because:

Row 1: 1 + 2 + 3 = 6
Row 2: 4 + 5 + 6 = 15
ðŸŽ¯ Memory Trick

axis=0 â†’ vertical â†“
axis=1 â†’ horizontal â†’
